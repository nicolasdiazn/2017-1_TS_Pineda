Se define una función que calcula cual el punto de la sucesión $x_{n+1} = r x_n (1-x_n)$ después de N pasos partiendo del punto $x_0$

```julia
function limite(r, x0, N=100)
    # Se inicializa x_n
    x_n = x0
    # Se repite N veces la sucesion 
    for i in 0:N
        # Se calcula x_{n+1}
        x_n1 = r*x_n*(1.0-x_n)
        # Se guarda la informacion del paso n+1 como si fuera el paso n para
        # volver aplicar el mismo ciclo
        x_n = x_n1
    end
    # Se regresa el valor x_N
    return x_n
end
```

```julia
# Se definen los limites en los que se van a barrer los valores de r
r_min = 0.0
r_max = 4.0

# Se define el numero de puntos 
N = 100000

# Se crea un arreglo vacio donde se guardaran los valores de x_N (r)
x = zeros(N)

# Se crea una "malla" uniforme para los valores de r
r = [(r_max-r_min)*(1.0*i)/N + r_min for i in 1:N]

# Se itera para cada r el valor de x_N dado un x_0 aleatoreo, con N = 1000<
for i in 1:N
    x[i] = limite(r[i], rand(), 1000)
end

# Se realiza un "scatter" de los valores de x_N en funcion de r

scatter(r, x, s=0.1)
xlabel("x")
ylabel("y")
xlim(r_min, r_max)
ylim(0,1);
```
