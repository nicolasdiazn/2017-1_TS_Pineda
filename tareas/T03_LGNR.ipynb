#### Se genera el espaciamiento. Ensamble de 100 matrices de 20x20

```julia
function matriz_aleatoria(n)
a=rand(n,n)+1im*rand(n,n);
b=triu(a,1);
c=transpose(conj(b))+b;
eigen=sort(eigvals(c));
espaciamiento=Array(Float64,size(eigen)[1]-1);
for i=2:size(eigen)[1]
        espaciamiento[i-1]=eigen[i]-eigen[i-1];
end
espaciamiento
end;
```

```julia
function ensamble(n)
    b=0;
for i=1:n
    a=matriz_aleatoria(20);
    b=cat(1,b,a);
end
b
end
```

```julia
datos=ensamble(100)
```

1. Calcular el centro de los puntos (como centro de masas)
2. Calcular la distancia de los puntos al centro
3. Elegir el punto más alejado al centro

4. Proponer distintas rutas con puntos diferentes y repitiendo sólo el de inicio (cerradas)
5. Calcular la longitud de cada ruta. La suma de la distancia entre cada par de puntos
6. Regresar al paso 4 hasta obtener todas las rutas cerradas y sin repeticiones posibles
7. Elegir la ruta con longitud más cercana a $2 \pi r$ dónde $r$ es la distancia del centro al punto más alejado escogido en el punto 3.
