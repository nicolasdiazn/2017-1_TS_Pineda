```julia
function logisticmap(N::Int,r,x0)
    V=Array(Float64, N)
    x=x0;
    for i in 1:N
       x=r*x*(1-x)
        V[i]=x
    end
    return V
end

function logisticmaps(N::Int,Ncut::Int,r,step)
    A=Array(Float64,N-Ncut,ceil(Int,(1-2*step)/step)+1)
    i=1;
    for x0=step:step:1-step
        A[:,i]=deleteat!(logisticmap(N,r,x0),1:Ncut);
        i=i+1;
    end
    return vec(A)
end

using PyPlot
function logisticplot(N::Int,Ncut::Int,rstep,xstep)
    rs=linspace(1,4,rstep)
    xs=[logisticmaps(N,Ncut,r,xstep) for r in rs]
    plot(rs,xs)
end
```

```julia
@time logisticplot(5000,4950,1000,.2)
```
